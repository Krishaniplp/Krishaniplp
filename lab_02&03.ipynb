```python
import pandas as pd
import geopandas
import matplotlib.pyplot as plt
```

```python
from google.colab import files
data_to_load = files.upload()
```

```python
df = pd.read_csv('Data_SL.csv', index_col=0)
df
```

```python
df['geometry'] = geopandas.GeoSeries.from_wkt(df['geometry'])
```

```python
geo_df = geopandas.GeoDataFrame(df, crs="EPSG:4326")
geo_df
```

```python
temp_today = [30,29,26,29,25,29,29,31,30]
geo_df['temp_today'] = temp_today
geo_df
```

```python
geo_df['population_density'] = geo_df['population']/geo_df['area']
geo_df
```

```python
geo_df.plot(column='population_density', legend = 'true', cmap = 'OrRd')
```

```python
geo_df.plot(column='rain_today', legend=True, cmap='plasma', legend_kwds={'label': "Rainfall Today : Sri Lanka"})
plt.show()
```
