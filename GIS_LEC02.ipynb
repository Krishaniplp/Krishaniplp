```python
import pandas as pd
import geopandas
import matplotlib.pyplot as plt
```

```python
df_new = pd.DataFrame(
{'City': ['Mumbai', 'Colombo', 'Bankok', 'Dhaka ', 'Chennai'],
'Country': ['India', 'Sri Lanka', 'Thailand', 'Bangladesh', 'India'],
'Latitude': [19.07, 6.92, 13.75, 23.81, 13.08],
'Longitude': [72.87, 79.86, 100.50, 90.41, 80.27]})
```

```python
df_new
```

```python
geo_df = geopandas.GeoDataFrame(df_new, geometry=geopandas.points_from_xy(df_new.Longitude, df_new.Latitude))
```

```python
geo_df
```

```python
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
world
```

```python
ax = world[world.continent=='Asia'].plot(color='white', edgecolor='black')
geo_df.plot(ax=ax, color='red')
plt.show()
```
